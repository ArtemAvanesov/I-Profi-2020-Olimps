```python
# Вам нужно реализовать алгоритм kNN — k Nearest Neighbours. На плоскости даны n
# точек разбитые на 10 классов (какие-то классы могут быть пустыми). Алгоритм kNN
# классифицирует произвольную точку (x,y) плоскости относя ее к одному из классов
# следующим образом: рассматриваются k ближайщих к ней точек и класс (x,y) относится
# к тому классу, который наиболее часто встречается среди этих k точек. (Если таких
# классов несколько, то берется минимальный по номер класс).
# Формат ввода
# В первой строке даны 3 натуральных числа, разделенные пробелами, — 
# n,k,t. 1≤k≤n≤1000. 1≤t≤100.
# На следующих n строчках заданы размеченные данные: в i-той строке 3 числа, разделенные
# пробелами, xi,yi,ci — координаты i-той точки и метка ее класса. xi,yi — действительные
# числа от −1000 до 1000, ci — натуральное число от 1 до 10.
# Затем на следующих t строчках заданы тестовые данные: в i-той строке 2 числа, разделенные
# пробелами, xi,yi — координаты i-той точки тестовых данных. xi,yi — действительные числа
# от −1000 до 1000.
# Формат вывода
# Вывод вашей программы должен состоять из 
# t строк, в i-той строке, должно находиться одно единственное число — ci метка класса i-той
# точки тестовых данных предсказанная с помощью kNN.
# Пример
# Ввод
# 5 3 3
# 1 0 1
# 2 0 2
# 3 0 1
# 4 0 1 
# 5 0 1
# 3 4
# 4 10
# 100 100
# Вывод
# 1
# 1
# 1

import sys

first_str = input()
aa = first_str.split(" ")

n = int(aa[0])
k = int(aa[1])
t = int(aa[2])

data = []

for i in range(0, n+t):
    loc=input().split(" ")
    data.append(loc)

train = []
test = []

for i in range(0, n):
    train.append([float(data[i][0]),float(data[i][1]),float(data[i][2])])

for i in range(n, n+t):
    test.append([float(data[i][0]),float(data[i][1])])

def euclideanDistance(point, train):
    distance = []
    for i in range(n):
        distance.append([((train[i][0]-point[0])**2 + (train[i][1]-point[1])**2), train[i][2]])
    return distance

def searh_k_min_distances_ret_class(distances_data, num_neigbours):
    sort_distances = sorted(distances_data)
    short_data = sort_distances[:num_neigbours]
    short_data
    classes = []
    for i in range(num_neigbours):
        classes.append(short_data[i][1])  
    return classes

def one_class(classes):
    classes = sorted(classes)
    most_common = None
    count_most_common = 0
    for item in classes:
        count = classes.count(item)
        if count > count_most_common:
            count_most_common = count 
            most_common = item
    return int(most_common)

answer = []
for i in range(t):
    distances = euclideanDistance(test[i], train)
    classes = searh_k_min_distances_ret_class(distances,k)
    class_point = one_class(classes)
    answer.append(class_point)

for item in answer:
    print(str(item))
    sys.stdout.flush()
```
