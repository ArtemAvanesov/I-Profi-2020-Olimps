```python
# При создании операционной системы реального времени для мобильного робототехнического
# устройства тестировщиками были проведены испытания разрабатываемого пакета программ,
# используя 16 вариантов наборов исходных данных. Варианты данных распределялись в
# соответствии с некоторой функцией частот, как показано в таблице. В 9 тестах обнаружены
# ошибки, которые не исправлялись. По результатам испытаний оценить надежность пакета
# программ с точки зрения вероятности отсутствия отказов. 
# № варианта Частота выбора варианта Результат теста
# 1             0,06                    0
# 2             0,02                    1
# 3             0,08                    0
# 4             0,06                    1
# 5             0,12                    0
# 6             0,02                    1
# 7             0,05                    0
# 8             0,07                    1
# 9             0,10                    0
# 10            0,05                    1
# 11            0,08                    0
# 12            0,08                    1
# 13            0,03                    0
# 14            0,07                    1
# 15            0,10                    0
# 16            0,01                    1

import numpy as np 

array = [[0.06, 0],
         [0.02, 1],
         [0.08, 0],
         [0.06, 1],
         [0.12, 0],
         [0.02, 1],
         [0.05, 0],
         [0.07, 1],
         [0.10, 0],
         [0.05, 1],
         [0.08, 0],
         [0.08, 1],
         [0.03, 0],
         [0.07, 1],
         [0.10, 0],
         [0.01, 1]]

value = 0
for i in range(16):
    if(array[i][1]==0):
        value += array[i][0]
print(value)

value2 = 0
for i in range(16):
    if(array[i][1]==1):
        value2 += array[i][0]
print(value2)
```
