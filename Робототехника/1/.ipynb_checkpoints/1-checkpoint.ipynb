```python
# Мобильный робот имеет на борту лазерный дальномер, имеющий точность 2 мм,
# и сонар, имеющий точность 20 мм. В момент остановки робота каждый из
# данных сенсоров совершает 6 измерений расстояния до ближайшего препятствия.
# Получаются следующие массивы значений:
# – измерения дальномера: 170, 167, 168, 176, 172, 165.
# – измерения сонара: 174, 182, 178, 170, 172, 178.

# Каково истинное расстояние до препятствия? Ответ введите в формате числа с
# точностью до второго знака после запятой.

import numpy as np 
import matplotlib.pyplot as plot

dalnometr = [170, 167, 168, 176, 172, 165]  # 2mm
sonar = [174, 182, 178, 170, 172, 178] # 20mm

print("Данные с дальнометра: ", dalnometr)
print("Данные с сонара: ", sonar)

dalnometr_with_prog_max = np.array(dalnometr)+2
dalnometr_with_prog_min = np.array(dalnometr)-2
print("Верхняя граница дальнометра с учетом погрешности: ", dalnometr_with_prog_max)
print("Нижняя граница дальнометра с учетом погрешности: ", dalnometr_with_prog_min)

sonar_with_prog_max = np.array(sonar)+20
sonar_with_prog_min = np.array(sonar)-20
print("Верхняя граница сонара с учетом погрешности: ", sonar_with_prog_max)
print("Нижняя граница сонара с учетом погрешности: ", sonar_with_prog_min)

print("max dal: ", max(dalnometr))
print("min dal: ", min(dalnometr))

print("max son: ", max(sonar))
print("min son: ", min(sonar))

#for i in range(len(dalnometr)):
print("Среднее по dal: ", sum(dalnometr)/6)
print("Среднее по son: ", sum(sonar)/6)
print("Среднее по средним: ", (sum(dalnometr)/6+sum(sonar)/6)/2)

plot.plot([1,2,3,4,5,6],dalnometr)
plot.plot([1,2,3,4,5,6],sonar)
plot.show()
```
